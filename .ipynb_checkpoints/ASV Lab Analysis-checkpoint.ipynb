# Anodic Stripping Voltammetry (ASV)

ASV is an analytical method to determine the concentration of metal ions in a solution. Its most notable quality is it's ability to detect very low concentrations; concentraions of metal ions as low as $10^{-10}$ M have been detected using this method. This particular experiment used a hanging drip mercury electrode (HDME) as its working electrode, there are various other viable working electrodes. The HDME is submerged into solution and the solution is then purged of all atmospheric gases. A cathodic potential is then applied to allow the deposition of metal ions into the mercury droplet. This depostion step can take anywhere from 1 minute to 30 minutes, depending on the concentraion of the solution. The HDME the goes through "quiet time" where stirring is stopped for about 30 seconds to allow the electrode to equlibriate. Finally, an linear anodic current is applied, stripping the HDME of the metal ions, and the current in the cell is recorded as a function of potential. At a certain potential (dependent of the metal ion being analyzed), the metal ions start to oxidize, causing a sharp increase in current. The height of the peak current is proportional to the weight of the deposited metal. 

Two analyses were performed using ASV: (1) determination of cadmium ions in an unknown solution using a calibration curve and (2) determination of a lead ions in an unknown solution using the standard addition method.

While ASV is a highly sensitive and accurate method of determining unknow concentrations of metals, the analysis of the .txt files produced is very time consuming and tedious to do through excel.

## Single analysis 
Now that we've located all the files that we want to analyze, lets extract the information we need from it starting one file at at time. Once we do this to one file we'll try to batch analyze them but lets keep it simple for now. 

From the data, we need to determine the magnitide of the peak current, $i_{p}$. To do this we need to find the minimum current at the oxidization potential, we'll call this $i_{min}$ and take a linear regression of the deposition time and the calculate a value we'll call $i_{max}$ (it's not exactly the max current but, for lack of a better description, this will do). Then we can calculate the $i_{p}$ by subtracting $i_{min}$ from $i_{max}$

This figure will help you vizualize what these values are


$i_{min}$ is the value of the current at the bottom of the anodic peak (ie. at oxidizing potential) and $i_{max}$ is the value from the depostion linear fit extrapolated to the oxidizing potential

```python
sample_fit = np.loadtxt(all_ASV_files[0], skiprows = 18, delimiter = ',', max_rows = 450) # the "all_ASV_files[0]" bit just gives us the first file from the asv files glob

sample_pot = sample_fit[:,0] # creates an array containing all elements in the 0th column of the data file 
sample_cur = sample_fit[:,1] # creates an array containing all elements in the 1st column of the data file

dep_pot_samp = sample_fit[:150, 0] # grabs first 150 potentail values 
dep_cur_samp = sample_fit[:150, 1] # grabs first 150 current values

x_fit_extended = np.arange(-800, -350, 1) # creates an array that will allow us to extrapolate the line out to all x values 

i_min_sample = min(sample_cur) # find the minimum of the y-vaules, the minimum current 
p_ox_sample = sample_pot[sample_cur.argmin()] # find the corresponding x-vaule for the y min, the oxidizing potential 


plt.plot(sample_pot, sample_cur)
m_sample, b_sample = np.polyfit(dep_pot_samp, dep_cur_samp, 1) # give slope and y-intercept of linear regresison of first 150 values. 
plt.plot(x_fit_extended, m_sample*x_fit_extended+b_sample, linestyle = 'dashed', label = 'deposition fit' ) # plots the extrapolated linear regression. 
plt.xlabel('Potential (mV)')
plt.ylabel('Current (A)')
plt.legend()
print(' Slope:', m_sample, '\n Y-intercept:', b_sample, '\n Minimum Current (A): ', i_min_sample, '\n Oxidizing Potential (mV): ', p_ox_sample) 

i_max_sample = p_ox_sample*m_sample + b_sample

plt.annotate("", xy=(p_ox_sample , i_max_sample), xycoords="data", xytext=(p_ox_sample, i_min_sample), textcoords="data", arrowprops=dict(arrowstyle="->", connectionstyle="arc3"))
plt.text(p_ox_sample+15, -0.000002, "$I_{p}$", fontsize='x-large')
```

## Batch analysis 

Now that we've figured out how to do a full analysis on one data file, lets try to load up every file we need at once and create a loop that will analyze them all in one go. 

Since we've already laid out the groundwork in the single analysis all we really have to do is to create a for loop, and copy the code we performed in the single analysis to the loop. We have to change what we call our arrays a little bit and the `np.loadtxt()` contents to accomadate but its a pretty quick process. 

Let's analyze the calibration curve data. 

```python
calib_files = glob.glob("?X10-5.txt*") #groups together only the files we need to construct the calibration curve 

i=1 # starts an iterator 
for calib_curve in calib_files:
    potential, current = np.loadtxt(calib_curve, skiprows = 18, delimiter = ',', max_rows = 450, usecols = (0,1), unpack = True)
    
    dep_pot = potential[:150]  
    dep_cur = current[:150]
    
    x_fit_extended = np.arange(-800, -350, 1)
    
    m_dep, b_dep = np.polyfit(dep_pot, dep_cur, 1)# give slope and y-intercept of linear regresison of first 150 values. in the polyfit(), 1 tells it to treat it as a linear fit 
    
    i_min = min(current) # find the minimum of the y-vaules, the minimum current 
    p_ox = potential[current.argmin()] # find the corresponding x-vaule for the y min, the oxidizing potential 
    
    i_max = m_dep*p_ox + b_dep
    i_p = i_max - i_min
    
    plt.figure(i)
    i=i+1 # creates an iterator so that each data file can have its own plot. 
    plt.plot(potential, current) 
    plt.plot(x_fit_extended, m_dep*x_fit_extended+b_dep, linestyle = 'dashed', label = 'deposition fit' ) # plots the extrapolated linear regression. 
    plt.xlabel('Potential (mV)')
    plt.ylabel('Current (A)')
    plt.title(f'{calib_curve} \n Oxidizing Potential (mV) = {p_ox} \n Peak Current(A) = {i_p} \n')
    plt.legend()
    plt.annotate("", xy=(p_ox , i_max), xycoords="data", xytext=(p_ox, i_min), textcoords="data", arrowprops=dict(arrowstyle="->", connectionstyle="arc3"))
    plt.text(p_ox+15, -0.5*i_p, "$I_{p}$", fontsize='x-large')
   
    #print(' Slope:', m, '\n Y-intercept:', b, '\n Minimum Current (A): ', i_min, '\n Oxidizing Potential (mV): ', p_ox, '\n Peak Current(A):', i_p, '\n')
    #^probably don't need this line but ill leave it here just in case
```

What happening here?

We start by creating an array `calib_files` that contains every file we need to construct the calibration curve. then we start an iterator loop with `i=1`, which we'll use and explain later. 

Then we start the 'for' loop. 'for _ in _' loops are used when we have a block of code that we want to repeat a fixed number of times. You feed the loop an array (of data files in this case), and then the loop perfoms whatever block of code you feed it to every member of the array, one-by-one, until the end of the array. So, simply put, when we write `for calib_curve in calib_files:` we're saying "***for the calibration curve, do 'this' to every member in the array 'calib_files'***"

We define what 'this' is in the following hanging code. The first thing we tell the code to do is to load the data from a file and place the values from the first column in an array named  'potential' and form the second column in an array called 'current'. Then we create a second set of arrays that contain only the first 150 values from each column in the data file. Next, we create an array that allows us to extend the deposition linear fit across the entirty of the plot. Then we find $i_{min}$ and the oxidizing potential. Finally, we plot the data set and the deposition linear fit over eachother and calculate $i_{p}$. 

The reason we set`i=1` in the begining is so that when we get to the `plt.figure()` part, the first plot will be declared figure 1, and then we add 1 to `i` in the line following so at the end of the first loop, `i` now equal 2. Then the loop starts again, declaring the next plot as figure 2 and `i` equal to three and so on. Thats how we got each data set on its own plot instead overlapping eachother like it would've done without this detail.

Now you can change out some of the details depending on what type of analysis you're doing, but this is the basic procedure you would use to create a for loop to batch analyze a set of data files: Declare for loop -> load data into arrays -> plot data -> comb through data arrays for whatever info you need. 

## Construct the Calibration Curve and Determine Unknown Concentration of $Cd^{2+}$

We need to figure out a way to take the $i_{p}$ and concentration of each run and compile them into respective arrays, plot the arrays and perform a linear regression on it. 

we could just manually insert each value into an array bacause they're all printed in the output of the cell above, but it's be more fun to find a way to automate this process. 

`np.append` is an easy way to accomplish this. We just need to create some empty arrays and then call the append function in the for loop. Once we do this we can simply plot the the two arrays and perform a linear regression on it. The equation of the line thats obtained from the equation will give us the relationship between $i_{p}$ and $[Cd^{2+}]$, allowing us to determine the concentration of the unknown solution. 

We'll need to use another for loop to automate this process.

```python
import scipy.stats

i_pArray = []
concArray = np.arange(0.00001, 0.00009, 0.00002)

for calib_curve in calib_files:
    potential, current = np.loadtxt(calib_curve, skiprows = 18, delimiter = ',', max_rows = 450, usecols = (0,1), unpack = True)
    
    dep_pot = potential[:150]  
    dep_cur = current[:150]
    m_dep2, b_dep2 = np.polyfit(dep_pot, dep_cur, 1)
    
    i_min = min(current)
    p_ox = potential[current.argmin()]  
    
    i_max = m_dep2*p_ox + b_dep2
    i_p = i_max - i_min
    
    i_pArray = np.append(i_pArray, i_p)
    
plt.plot(concArray, i_pArray, 'ro')
plt.xlabel('Concentration (M)')
plt.ylabel('Peak Current (A)')

#linear Regression
m_fit, b_fit = np.polyfit(concArray, i_pArray, 1)
line = np.arange(0.00001, 0.00007, 0.000001)
plt.plot(line, m_fit*line+b_fit, linestyle = 'dashed')
stats = scipy.stats.linregress(concArray, i_pArray)

plt.title(f' Calibration Curve of Standard Cadmium Solutions \n  i_p = {m_fit}C + {b_fit} \n  Coefficient of Determination = {stats[2]} \n')


#print(' Peak Current Values(A) :', i_pArray, '\n Concentrations (M): ', concArray)
#print(' slope:', m_fit, '\n y-intercept:', b_fit, '\n coefficient of determination:', stats[2] )
```

Now we need to load our data for the unknown trials, determine the $i_{p}$ values from each trial, calculate the average $i_{p}$ and then calculate the Concentraion of the unknown. We can pretty much just copy and paste the for loop again and then add some math at the end.  

```python
Cd_unk_data = glob.glob("CD?_UNK.txt*")

i_pArray_unkCd = []

for unk_Cd in Cd_unk_data:
    potential_unkCd, current_unkCd = np.loadtxt(unk_Cd, skiprows = 18, delimiter = ',', max_rows = 450, usecols = (0,1), unpack = True)
    
    dep_pot_unkCd = potential_unk_Cd[:150]  
    dep_cur_unkCd = current_unk_Cd[:150]
    m_depCd, b_depCd = np.polyfit(dep_pot_unkCd, dep_cur_unkCd, 1)
    
    i_min_unkCd = min(current_unkCd)
    p_ox_unkCd = potential_unkCd[current_unkCd.argmin()]  
    
    i_max_unkCd = m_depCd*p_ox_unkCd + b_depCd
    i_p_unkCd = i_max_unkCd - i_min_unkCd
    
    i_pArray_unkCd = np.append(i_pArray_unkCd, i_p_unkCd)

avg_i_p_Cd = np.average(i_pArray_unkCd)
print(' Average Peak Current of Cd Unkowns: ', avg_i_p_Cd)
conc_Cd = (avg_i_p_Cd - b_fit) / m_fit
print('\n Concentraion of Cd Unkown: ', conc_Cd)
```
